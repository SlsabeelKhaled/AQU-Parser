#  The main function of the Parser

```python
from lex import *
from parse import *
import sys

def main():
    print("Teeny Tiny Compiler ")
    input_file = input("Enter File Name :")
    file = open(input_file, 'r')
    
    
    # Initialize the lexer and parser.
    lexer = Lexer(file.read())
    parser = Parser(lexer)
    parser.program() 
    
    print("Parsing completed.")

main()
```
